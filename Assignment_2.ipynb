```python
import yfinance as yf
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import time
```

```python
#Decorator to check performance
def performance(func):
  def wrapper(*args, **kwargs):
    start = time.time()
    start_time_str = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S:%f")
    result = func(*args, **kwargs)
    end = time.time()
    print("----->Starting time: ", start_time_str)
    print("----->Ending time: ", datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S:%f"))
    print("----->",func.__name__ + " took " + str((end - start)*1000) + " milli sec")
  return wrapper
```

```python
ticker = input("Enter the Stock ticker: ")
SBUX_ticker = yf.Ticker(ticker)
start_date = datetime.date(2021, 1, 1)
end_date = datetime.date.today()
period = start_date-end_date
data = SBUX_ticker.history(start=start_date)
data
```

```python
daily_investment = 6.5
weekly_investment = 19.5
monthly_investment = 4.2 * 3 * 6.5
investment_days = ["Monday", "Wednesday", "Friday"]
monthly_investment_dates = []

#Daily_Investment
@performance
def calculate_daily_investment(opinion):
  data["Daily_investment_closed"] = daily_investment / data["Close"]
  data["Daily_investment_opened"] = daily_investment / data["Open"]
  if(opinion == "Close"):
    print("Skiraks worth if he invest daily: ",data[data["Weekday"].isin(investment_days)]["Daily_investment_closed"].sum())
  elif(opinion == "Open"):
    print("Skiraks worth if he invest daily: ",data[data["Weekday"].isin(investment_days)]["Daily_investment_closed"].sum())

#Weekly_Investment
@performance
def calculate_weekly_investment(opinion):
  data["Weekly_investment_closed"] = weekly_investment / data["Close"]
  data["Weekly_investment_opened"] = weekly_investment / data["Open"]
  if(opinion == "Close"):
    print("Srikars worth if he invest weekly: ", data[data["Weekday"] == "Monday"]["Weekly_investment_closed"].sum())
  elif(opinion == "Open"):
    print("Srikars worth if he invest weekly: ", data[data["Weekday"] == "Monday"]["Weekly_investment_opened"].sum())

#Monthly_Investment
@performance
def calculate_monthly_investment(opinion):
  data["Monthly_investment_closed"] = monthly_investment / data["Close"]
  data["Monthly_investment_opened"] = monthly_investment / data["Open"]
  month_end_dates = pd.date_range(start = start_date, end = end_date, freq = 'ME')
  for month in month_end_dates:
    investment_date = month
    if month.strftime("%A") in "Saturday":
      investment_date = month - datetime.timedelta(1)
    elif month.strftime("%A") in "Sunday":
      investment_date = month - datetime.timedelta(2)
    monthly_investment_dates.append(investment_date.date())
  if(opinion == "Close"):
    print("Srikars worth if he invests monthly: ", data[data.index.isin(monthly_investment_dates)]["Monthly_investment_closed"].sum())
  elif(opinion == "Open"):
    print("Srikars worth if he invests monthly: ", data[data.index.isin(monthly_investment_dates)]["Monthly_investment_opened"].sum())


calculate_daily_investment("Close")
calculate_weekly_investment("Close")
calculate_monthly_investment("Close")
data
```
