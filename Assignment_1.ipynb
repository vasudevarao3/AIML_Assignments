#Question - 1#

```python
ticker = input("Enter the stock ticker:")
```

```python
#data = yf.download(ticker, start=datetime.datetime.today()-datetime.timedelta(30), end=datetime.datetime.today())
ticker = yf.Ticker(ticker)
data = ticker.history(period="1y")
```

```python
print(data)
```

```python
seven_days_data = data.tail(7)
seven_days_data
```

```python
type(seven_days_data)
```

```python
closing_list = seven_days_data.iloc[:,1].tolist()
closing_list
```

```python
sum_closing_list = sum(closing_list)
sum_closing_list
```

```python
avg_closing_list = sum_closing_list/len(closing_list)
print("Average of Closing List of Last 7 Days: ",avg_closing_list)
print("Closing list prices: ", closing_list)
```

#Question - 3#


```python
ticker = input("Enter the ticker symbol: ")
ticker = yf.Ticker(ticker)
data = ticker.history(period="6mo")
data
```

```python
data.describe()
```

```python
data["Daily Return"] = data["Close"].pct_change()
data
```

```python
data.describe()
```

```python
print("Mean : ", data["Daily Return"].mean())
print("Stadard Deviation: ", data["Daily Return"].std())
print("Minimum : ", data["Daily Return"].min())
print("Maximum : ", data["Daily Return"].max())
```

```python
import matplotlib.pyplot as plt
plt.hist(data["Daily Return"])
plt.show()
```
